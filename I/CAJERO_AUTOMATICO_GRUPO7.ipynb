```python
import random

def menu():  # FUNCION MENU. RECIBE LA OPCIÓN DISPONIBLE PARA REALIZAR
    option = 0
    print("\n1.Consultar saldo\n2.Depositar\n3.Retirar\n4.Transferir\n5.Salir\n")
    option = int(input("---> "))
    print("-"*46)
    while option < 1 or option > 5:
        print("-"*46)
        print("\nINGRESA UNA OPCIÓN CORRECTA*\n\n1.Consultar saldo\n2.Depositar\n3.Retirar\n4.Transferir\n5.Salir\n")
        option = int(input("---> "))
        print("-"*46)
    return option


def consultarSaldo(saldo):  # FUNCION QUE IMPRIME EL SALDO YA GENERADO ALEATORIAMENTE
    print("Tu saldo es: $", saldo)
    print("-"*46)



def depositar(saldo):  # FUNCION PARA AGREGAR VALORES SIMULADOS AL SALDO QUE SE DEPOSITEN EN EL CAJERO
    deposito = int(input("¿Cuánto desea depositar?\n---> "))
    confirmacion = int(
        input("\n¿Desea confirmar el deposito?\n1. SI\n2. NO\n---> "))

    if confirmacion != 1:
        print("\n¡Deposito cancelado!\n")
        return saldo
    else:
        print("¡Depósito exitoso!")
        saldo = saldo + deposito
        print("Tu nuevo saldo es: ", saldo)
        print("-"*46)
        return saldo


def retirar(saldo):  # SE RETIRA EL DINERO QUE EL USUARIO PIDA MIENTRAS TENGA FONDOS

    retiro = int(input("¿Cuánto desea retirar? RETIRO MÍNIMO $10000*\n---> "))
    while True:
        if (retiro < saldo and retiro >= 10000):
            break
        else:
            retiro = int(input("\nFondos insuficientes o retiro menor del mínimo permitido*\n\n¿Cuánto desea retirar? RETIRO MÍNIMO $10000*\n---> "))

    saldo = saldo - retiro
    print("\n¡Retiro exitoso!, tu nuevo saldo es: ", saldo)
    print("-"*46)
    return saldo


def transferir(saldo):
    input("Número de cuenta a transferir\n--->")

    transferencia = int(input("\n¿Cuánto desea transferir?\n---> "))
    while transferencia > saldo:
        transferencia = int(input("\nFondos insuficientes \n\n¿Cuanto desea transferir?\n---> "))

    saldo = saldo - transferencia
    print("\nTransferencia exitosa, tu nuevo saldo es: $", saldo)
    print("-"*46)

    return saldo


print("------------ ¡Bienvenido! ------------")
option = 0
# EL RANGO DE VALORES DE LOS SALDOS, NO SUPE CON QUE VALORES GUIARME Y LO PUSE ASI
saldo = random.randrange(100000, 1000000, 5)

while option != 5:
    option = menu()

    if option == 1:
        consultarSaldo(saldo)
    elif option == 2:
        saldo = depositar(saldo)
    elif option == 3:
        saldo = retirar(saldo)
    elif option == 4:
        saldo = transferir(saldo)
else:
    print("¡Hasta luego!")
```
